# Open Government Data, Canton Zurich

### **Dataset**: Steuerstatistik zu juristischen Personen im Kanton Zürich

### **Description**: Die Steuerstatistik zu den juristischen Personen umfasst Statistiken zu den Steuerpflichtigen, Steuerfaktoren (Reingewinn/Kapital) und Steuererträgen nach Rechtsform, Gewinn- und Kapitalklassen wie auch Bezirk und Gemeinde. Bei den Steuererträgen handelt es sich um die einfache Staatssteuer, wenn nichts anderes angegeben ist. Die Daten stammen vom Kantonalen Steueramt. Gewinnsteuer und Kapitalsteuer werden jedes Jahr erhoben. Für die Bemessung der Gewinn-/Kapitalsteuer wird auf den Gewinn/das Kapital an einem bestimmten Stichtag (Gewinn/Kapital am Ende des Geschäftsjahres) abgestellt.
