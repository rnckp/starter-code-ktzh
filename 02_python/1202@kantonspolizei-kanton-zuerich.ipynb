# Open Government Data, Canton Zurich

### **Dataset**: Polizeiliche Kriminalstatistik (PKS) des Kantons Zürich

### **Description**: Die polizeiliche Kriminalstatistik (PKS) des Kantons Zürich wird jährlich publiziert. Sie beinhaltet Zahlen zu den registrierten Straftaten (und zu polizeilich beschuldigten sowie geschädigten Personen) für das Strafgesetzbuch (StGB), das Ausländer- und Integrationsgesetz (AIG) und das Betäubungsmittelgesetz (BetmG). Die polizeiliche Kriminalstatistik verzeichnet nur die polizeilich bekannt gewordene Kriminalität - das sogenannte Hellfeld. Sie ist eine Ausgangsstatistik. D.h. die Straftaten erscheinen nicht zum Tatzeitpunkt in der Statistik. Ausschlaggebend für die Zählung in der PKS ist der Zeitpunkt, an dem die Polizei einen ersten Rapport zur Straftat abgeschlossen hat.
